```r
zamienNaSystemy2Do16 <- function (liczbaDoZamiany) {
#     sprawdzamy, czy liczba, którą przekazaliśmy do zamiany jest co najmniej równa 0
    if (liczbaDoZamiany < 0) {
        print ("Liczba powinna być większa od 0.")
    } else {
        cat(paste(c("Liczba w zapisie dziesiętnym: ", liczbaDoZamiany)))
        cat('\n\n')
#         urachamiamy pętlę, która zamienia nam daną liczbę na systemy od 2 do 16
        for (i in 2:16) {
#             przypisujemy daną liczbę do zmiennej tymczasowej tmp, aby na niej operować a nadal pamiętać początkową przekazaną wartość
        tmp <- liczbaDoZamiany
#             używamy zmiennej "indeks" i wektora "liczbaWDanymSystemie" by przechowywać reszty z dzielenia danej liczby przez podstawe systemu liczbowego
        indeks <- 1
        liczbaWDanymSystemie = c()
#             nie kontynuujemy obliczeń jeśli przekażemy liczbę mniejszą od 16, czyli np. przekazujemy 10 to nie obliczamy jaką liczbą jest w systemie powyżej 10 (będzie to 0)
        if (liczbaDoZamiany >= i) {
            while (tmp > 0) {
            resztaZDzielenia = tmp %% i
#                 jeśli reszta z dzielenia jest większa od 9 musimy zamienić ją na literę
            if (resztaZDzielenia > 9) {
                resztaZDzielenia = zamienLiczbeDwucyfrowaNaLitere(resztaZDzielenia)
            }
#                 zapisujemy daną resztę z dzielenia do wynikowego wektora
            liczbaWDanymSystemie[[indeks]] = resztaZDzielenia
#                 dzielimy liczbę przez podstawę systemu i zaokrąglamy w dół
            tmp <- floor(tmp / i)
#                 zwiększamy indeks, pod którym będziemy zapisywać resztę z dzielenia w wektorze wynikowym
            indeks <- indeks + 1
            }
        }
#             wypisujemy funkcją cat() zamienioną liczbę na dany system liczbowy
#             funkcja paste() łączy nam elementy z wektora, zaś funkcja rev() odwraca nam kolejność wektora, np. 123 w 321
#             jest to spowodowane zależnościami w obliczeniach danej liczby w podanym systemie liczbowym
        cat(paste(i, paste(rev(liczbaWDanymSystemie), collapse=''), sep=': '))
        cat('\n\n')
        }
    }
}

zamienLiczbeDwucyfrowaNaLitere <- function (liczbaDwucyfrowa) {
#     zamieniamy przekazaną liczbę na liczbę, ale o typie tekstowym
#     robimy to, by użyć bloku switch - niestety wartość, po której wybieramy nie może być liczbą
    liczbaDucyfrowaJakoString <- paste(c(liczbaDwucyfrowa))
    switch(liczbaDucyfrowaJakoString, 
           '10' = 'A',
           '11' = 'B',
           '12' = 'C',
           '13' = 'D',
           '14' = 'E',
           '15' = 'F'
        )
}

# sprawdzamy działanie naszej funkcji
zamienNaSystemy2Do16(-1)
cat('\n\n\n\n')
zamienNaSystemy2Do16(10)
cat('\n\n\n\n')
zamienNaSystemy2Do16(25)
cat('\n\n\n\n')
zamienNaSystemy2Do16(1000)
cat('\n\n\n\n')
zamienNaSystemy2Do16(1294910)
```
