```r
znajdzNajwiekszyWspolnyDzielnik <- function(x, y) {
# wybieramy mniejszą z liczb, by nie szukać największego wspólnego dzielnika, który miałby być większy od którejś z liczb
    if(x > y) {
        mniejszaLiczba = y
    } else {
        mniejszaLiczba = x
    }
# iterujemy po liczbach całkowitych od 1 do mniejszej i sprawdzamy, czy obie te liczby są podzielne, przez tę z iteracji
    for(i in 1:mniejszaLiczba) {
        if((x %% i == 0) && (y %% i == 0)) {
            najwiekszyWspolnyDzielnik = i
        }
    }

    najwiekszyWspolnyDzielnik
}

# wypisujemy na ekran kilka przykładów wywołań napisanej funkcji
print(paste("Największy wspólny dzielnik dla", 10,"i", 15,"to", znajdzNajwiekszyWspolnyDzielnik(10, 15)))
print(paste("Największy wspólny dzielnik dla", 20,"i", 66,"to", znajdzNajwiekszyWspolnyDzielnik(20, 66)))
print(paste("Największy wspólny dzielnik dla", 13,"i", 999,"to", znajdzNajwiekszyWspolnyDzielnik(12, 999)))
print(paste("Największy wspólny dzielnik dla", 815,"i", 335,"to", znajdzNajwiekszyWspolnyDzielnik(815, 335)))
print(paste("Największy wspólny dzielnik dla", 444,"i", 572,"to", znajdzNajwiekszyWspolnyDzielnik(444, 572)))
print(paste("Największy wspólny dzielnik dla", 432,"i", 678,"to", znajdzNajwiekszyWspolnyDzielnik(432, 678)))
print(paste("Największy wspólny dzielnik dla", 293,"i", 789,"to", znajdzNajwiekszyWspolnyDzielnik(293, 789)))
print(paste("Największy wspólny dzielnik dla", 505,"i", 101,"to", znajdzNajwiekszyWspolnyDzielnik(505, 101)))
cat('\n\n\n\n')


# przepisujemy podaną wyżej funkcję na taką z użyciem algorytmu Euklidesa
znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa <- function(x, y) {
    while(y) {
        zmiennaPomocznicza = y
        y = x %% y
        x = zmiennaPomocznicza
    }
    
    x
}

# wypisujemy na ekran kilka przykładów wywołań nowo napisanej funkcji
print(paste("Największy wspólny dzielnik dla", 10,"i", 15,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(10, 15)))
print(paste("Największy wspólny dzielnik dla", 20,"i", 66,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(20, 66)))
print(paste("Największy wspólny dzielnik dla", 13,"i", 999,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(12, 999)))
print(paste("Największy wspólny dzielnik dla", 815,"i", 335,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(815, 335)))
print(paste("Największy wspólny dzielnik dla", 444,"i", 572,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(444, 572)))
print(paste("Największy wspólny dzielnik dla", 432,"i", 678,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(432, 678)))
print(paste("Największy wspólny dzielnik dla", 293,"i", 789,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(293, 789)))
print(paste("Największy wspólny dzielnik dla", 505,"i", 101,"to", znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(505, 101)))
cat('\n\n\n\n')


znajdzNajmniejszaWspolnaWielokrotnosc <- function(x, y) {
    # szukamy większej z liczb
    if(x > y) {
        wiekszaLiczba = x
    } else {
        wiekszaLiczba = y
    }
    
    # większą z liczb zwiększamy o 1 przy każdej iteracji oraz sprawdzamy, czy obie, podane na wejściu liczby, są dzielnikami
    # otrzymanej w pętli liczby
    # jeśli tak, kończym pętlę i zwracamy znalezioną liczbę
    while(TRUE) {
        if((wiekszaLiczba %% x == 0) && (wiekszaLiczba %% y == 0)) {
            najmniejszaWspolnaWielokrotnosc = wiekszaLiczba
            break
        }
        wiekszaLiczba = wiekszaLiczba + 1
    }
    
    najmniejszaWspolnaWielokrotnosc
}

# wypisujemy na ekran kilka przykładów wywołań napisanej funkcji
print(paste("Najmniejsza wspólna wielokrotność", 10,"i", 15,"to", znajdzNajmniejszaWspolnaWielokrotnosc(10, 15)))
print(paste("Najmniejsza wspólna wielokrotność", 20,"i", 66,"to", znajdzNajmniejszaWspolnaWielokrotnosc(20, 66)))
print(paste("Najmniejsza wspólna wielokrotność", 13,"i", 999,"to", znajdzNajmniejszaWspolnaWielokrotnosc(12, 999)))
print(paste("Najmniejsza wspólna wielokrotność", 815,"i", 335,"to", znajdzNajmniejszaWspolnaWielokrotnosc(815, 335)))
print(paste("Najmniejsza wspólna wielokrotność", 444,"i", 572,"to", znajdzNajmniejszaWspolnaWielokrotnosc(444, 572)))
print(paste("Najmniejsza wspólna wielokrotność", 432,"i", 678,"to", znajdzNajmniejszaWspolnaWielokrotnosc(432, 678)))
print(paste("Najmniejsza wspólna wielokrotność", 293,"i", 789,"to", znajdzNajmniejszaWspolnaWielokrotnosc(293, 789)))
print(paste("Najmniejsza wspólna wielokrotność", 505,"i", 101,"to", znajdzNajmniejszaWspolnaWielokrotnosc(505, 101)))
cat('\n\n\n\n')


znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc <- function(x, y) {
    # liczba1 * liczba2 = najmniejszaWspolnaWielokrotnosc * najwiekszyWspolnyDzielnik
    # najmniejszaWspolnaWielokrotnosc = liczba1 * liczba2 / najwiekszyWspolnyDzielnik
    
    najwiekszyWspolnyDzielnik = znajdzNajwiekszyWspolnyDzielnikMetodaEuklidesa(x, y)
    
    x * y / najwiekszyWspolnyDzielnik
}

# wypisujemy na ekran kilka przykładów wywołań napisanej funkcji
print(paste("Najmniejsza wspólna wielokrotność", 10,"i", 15,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(10, 15)))
print(paste("Najmniejsza wspólna wielokrotność", 20,"i", 66,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(20, 66)))
print(paste("Najmniejsza wspólna wielokrotność", 13,"i", 999,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(12, 999)))
print(paste("Najmniejsza wspólna wielokrotność", 815,"i", 335,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(815, 335)))
print(paste("Najmniejsza wspólna wielokrotność", 444,"i", 572,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(444, 572)))
print(paste("Najmniejsza wspólna wielokrotność", 432,"i", 678,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(432, 678)))
print(paste("Najmniejsza wspólna wielokrotność", 293,"i", 789,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(293, 789)))
print(paste("Najmniejsza wspólna wielokrotność", 505,"i", 101,"to", znajdzNajmniejszaWspolnaWielokrotnoscZnajacMatematycznaZaleznosc(505, 101)))
```
